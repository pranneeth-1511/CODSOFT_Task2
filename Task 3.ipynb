```python
#import the necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

#read the dataset file
dataset = pd.read_csv("/content/TASK 3 IRIS FLOWER CLASSIFICATION.csv")

#display the first five rows in the dataset
dataset.head(5)
```

```python
#to get the total information about the dataset
dataset.info()
```

```python
#to find the null value cell in respective columns
dataset.isnull().sum()
```

```python
#to get the total count of each species
dataset['species'].value_counts()
```

```python
#split the dataset into features and labels

x = dataset.drop('species',axis=1) #feature
y = dataset['species'] #species
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)

#get the shape of the axis
print("shape of X-axis:",x_train.shape)
print("shape of Y-axis:",y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression()

model.fit(x_train,y_train)
```

```python
#predcting the model of the dataset
y_pred=model.predict(x_test)

y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

#confusion mareix of the dataset
confusion_matrix(y_test,y_pred)
```

```python
#the accuracy of the given dataset
accuracy = accuracy_score(y_test,y_pred)*100
print("Accuracy of the model is",format(accuracy))
```
